```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
#RENO_data=pd.read_table('reno_data_cov')
data_cl68 = pd.read_table('../reno_for_global_analysis/reno_data/reno2018_cl68.csv', delim_whitespace=True)
data_cl95 = pd.read_table('../reno_for_global_analysis/reno_data/reno2018_cl95.csv', delim_whitespace=True)
data_cl99 = pd.read_table('../reno_for_global_analysis/reno_data/reno2018_cl99.csv', delim_whitespace=True)

RENO_cr_68=pd.read_table('reno_confidence_region_cov_68.dat',names=['s2','dm'],delim_whitespace=True)
RENO_cr_95=pd.read_table('reno_confidence_region_cov_95.dat',names=['s2','dm'],delim_whitespace=True)
RENO_cr_99=pd.read_table('reno_confidence_region_cov_99.dat',names=['s2','dm'],delim_whitespace=True)


```

# Región de confianza para RENO con el análisis $\chi^2_{cov}$

```python
data_RENO=pd.read_table('reno_data_cov.dat',names=['s2','dm','ji'],delim_whitespace=True)
```

```python
min_values=data_RENO.iloc[data_RENO['ji'].idxmin()]
s2_min=min_values[0]
dm_min=min_values[1]
ji_min=min_values[2]
data_RENO['ji']=data_RENO['ji']-ji_min
s2_min,dm_min,ji_min
```

```python
data_68=data_RENO[(data_RENO['ji'] <= 2.3) & (data_RENO['ji'] >= 2.25)]
data_95=data_RENO[(data_RENO['ji'] <= 6.18) & (data_RENO['ji'] >= 6.10)]
data_99=data_RENO[((data_RENO['ji'] <= 11.83) & (data_RENO['ji'] >= 11.71))]

data_68=data_68[['s2','dm']]
data_95=data_95[['s2','dm']]
data_99=data_99[['s2','dm']]
```

```python
fig, ax = plt.subplots()
ax.scatter(data_cl68['s22t13'].to_numpy(),data_cl68['dmee'].to_numpy(),label='99% of C.L. from RENO')
ax.scatter(data_cl95['s22t13'].to_numpy(),data_cl95['dmee'].to_numpy(),label='95% of C.L. from RENO')
ax.scatter(data_cl99['s22t13'].to_numpy(),data_cl99['dmee'].to_numpy(),label='68% of C.L. from RENO')
ax.scatter(data_99['s2'].to_numpy(),data_99['dm'].to_numpy(),s=0.5,c='red',label=r'99% of C.L. from $\chi^2_{cov}$')
ax.scatter(data_95['s2'].to_numpy(),data_95['dm'].to_numpy(),s=0.5,c='darkred',label=r'95% of C.L. from $\chi^2_{cov}$')
ax.scatter(data_68['s2'].to_numpy(),data_68['dm'].to_numpy(),s=0.5,c='black',label=r'68% of C.L. from $\chi^2_{cov}$')
ax.scatter(s2_min,dm_min,marker='P',c='black',label='Best fit from $\chi^2_{cov}$')

ax.set_xlim(0,0.2)
ax.set_ylim(0.0017,0.0035)
ax.set_xticks([0,0.05,0.10,0.15,0.20])
ax.set_yticks([0.002,0.0025,0.003,0.0035])
ax.set_xticklabels(['$0$','$0.05$','$0.10$','$0.15$','$0.20$'])
ax.set_yticklabels(['$2.0$','$2.5$','$3.0$','$3.5$'])

lg=ax.legend()

# Plot legend.
lgnd = plt.legend(loc="upper right", scatterpoints=1, fontsize=8)
lgnd.legendHandles[3]._sizes = [30]
lgnd.legendHandles[4]._sizes = [30]
lgnd.legendHandles[5]._sizes = [30]
lgnd.legendHandles[6]._sizes = [30]

ax.set(xlabel=r'$\sin^2 2\theta_{13}$', ylabel=r'$\Delta m^2_{ee} \hspace{0.5} [eV^{2}] \hspace{0.5} (x10^{-3})$')
ax.grid(True, linestyle='-')
ax.tick_params(labelcolor='black', labelsize=10, width=1)
fig.savefig("reno_cov.eps")
plt.show()
```

# Perfiles de la ji-cuadrada con matriz de covarianza para el experimento RENO

```python
s2_profile=data_RENO[data_RENO['dm']==dm_min]
dm_profile=data_RENO[data_RENO['s2']==s2_min]

s2_profile.to_csv('reno_s2_cov.dat')
dm_profile.to_csv('reno_dmee_cov.dat')
```

```python
# Datos del análisis de RENO
parabol_s2 = pd.read_table('parabol_s2_reno.dat', names=['s22t13','ji'],delim_whitespace=True)
parabol_dmee = pd.read_table('parabol_dmee_reno.dat', names=['dmee','ji'],delim_whitespace=True)

ji_min_s2=parabol_s2['ji'].min()
ji_min_dm=parabol_dmee['ji'].min()

parabol_s2['ji']=parabol_s2['ji']-ji_min_s2
parabol_dmee['ji']=parabol_dmee['ji']-ji_min_dm
```

### Perfil para $\sin^2 2\theta_{13}$

```python
fig, ax = plt.subplots()

ax.plot(s2_profile['s2'].to_numpy(),s2_profile['ji'].to_numpy(),c='black',linestyle='-.',label=r'Results from $\chi^2_{cov}$')
ax.plot(parabol_s2['s22t13'].to_numpy(),parabol_s2['ji'].to_numpy(),linestyle='--',label=r'RENO results')

ax.legend()
ax.set_xlim(0.00, 0.2)
ax.set_ylim(0.0, 6.5)
ax.grid(True, linestyle='-')

ax.set_xticks([0,0.05,0.1,0.15,0.2])
ax.set_yticks([1,2,3,4,5,6])
ax.set_xticklabels(['0','$0.05$','0.10','$0.15$','0.2'])
ax.set_yticklabels(['','$2$','','$4$','','$6$',''])

ax.set(xlabel=r'$\sin^2 2\theta_{13}$', ylabel=r'$\Delta \chi^2$')

fig.savefig("reno_parabol_s22t13_cov.eps")
```
