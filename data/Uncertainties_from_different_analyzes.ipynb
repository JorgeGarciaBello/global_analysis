```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# example data


fig, ax0 = plt.subplots(figsize=(9,6))
#ax0.errorbar(x,y,xerr=sigma_err,fmt='o')
# RENO results
ax0.errorbar(0.0896,7,xerr=[[0.00671],[0.00671]],fmt='o')
# Our RENO results
ax0.errorbar(0.0896,6,xerr=[[0.00712],[0.00719]],fmt='o')

# Daya Bay results
ax0.errorbar(0.0856,5,xerr=[[0.00305],[0.00288]],fmt='o')
# Our Daya Bay results
ax0.errorbar([0.0856],[4],xerr=[[0.0029],[0.0029]],fmt='o')

# Our RENO and Daya Bay results
ax0.errorbar([0.0873],[3],xerr=[[0.00267],[0.00265]],fmt='o')



# Acotaciones en los ejes X y Y
ax0.set_xlim(0.08, 0.1)       # From RENO
ax0.set_ylim(2.5, 7.5)           # From RENO


ax0.set_xlabel(r'$\sin^2 2 \theta_{13}$')
ax0.set_ylabel('Best fit per analysis')

ax0.set_xticks([0.08,0.085,0.09,0.095,0.1])
ax0.set_xticklabels(['0.080','0.085','$0.090$','0.095','0.100'])

ax0.set_yticks([3,4,5,6,7])
ax0.set_yticklabels(['','','','',''])
ax0.margins(2)           # Default margin is 0.05, value 0 means fit
# Añado leyenda, tamaño de letra 10, en esquina superior derecha
ax0.legend(('RENO',r'$\chi^2_{RENO}$','Daya Bay', r'$\chi^2_{DB}$',r'$\chi^2_{RENO+DB}$' ),
prop = {'size': 15}, loc='lower right')

ax0.xaxis.label.set_size(18)
ax0.yaxis.label.set_size(20)
ax0.tick_params(labelcolor='black', labelsize=15, width=1)

fig.savefig("s22t13_uncertainties_comparison.png")
#fig.savefig("s22t13_uncertainties_comparison.eps")
```
