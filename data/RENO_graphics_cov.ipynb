```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
#RENO_data=pd.read_table('reno_data_cov')
data_cl68 = pd.read_table('../reno_for_global_analysis/reno_data/reno2018_cl68.csv', delim_whitespace=True)
data_cl95 = pd.read_table('../reno_for_global_analysis/reno_data/reno2018_cl95.csv', delim_whitespace=True)
data_cl99 = pd.read_table('../reno_for_global_analysis/reno_data/reno2018_cl99.csv', delim_whitespace=True)

RENO_cr_68=pd.read_table('reno_confidence_region_cov_68.dat',names=['s2','dm'],delim_whitespace=True)
RENO_cr_95=pd.read_table('reno_confidence_region_cov_95.dat',names=['s2','dm'],delim_whitespace=True)
RENO_cr_99=pd.read_table('reno_confidence_region_cov_99.dat',names=['s2','dm'],delim_whitespace=True)

```

```python
RENO_min_cov=pd.read_table('min_param_reno_data_cov.dat',names=['ji','s2','dm'],delim_whitespace=True)
s2_min=RENO_min_cov['s2'].iloc[0]
dm_min=RENO_min_cov['dm'].iloc[0]
s2_min,dm_min
```

```python
fig,ax=plt.subplots()
ax.scatter(data_cl68['s22t13'].to_numpy(),data_cl68['dmee'].to_numpy(),label='68% of C.L. from RENO')
ax.scatter(data_cl95['s22t13'].to_numpy(),data_cl95['dmee'].to_numpy(),label='95% of C.L. from RENO')
ax.scatter(data_cl99['s22t13'].to_numpy(),data_cl99['dmee'].to_numpy(),label='99% of C.L. from RENO')
ax.scatter(RENO_cr_68['s2'].to_numpy(),RENO_cr_68['dm'].to_numpy(),s=0.1,c='black',label=r'68% of C.L. from $\chi^2_{cov}$')
ax.scatter(RENO_cr_95['s2'].to_numpy(),RENO_cr_95['dm'].to_numpy(),s=0.2,c='darkred',  label='95% of C.L. from $\chi^2_{cov}$')
ax.scatter(RENO_cr_99['s2'].to_numpy(),RENO_cr_99['dm'].to_numpy(),s=0.3,c='red', label='99% of C.L. from $\chi^2_{cov}$')
ax.scatter(s2_min,dm_min,c='black',marker='+' ,label='Our best fit')
# Limites en los ejes
ax.set_xlim(0.0,0.2)
ax.set_ylim(0.002,0.0035)

ax.set_xticks([0,0.05,0.1,0.15,0.2])
ax.set_yticks([0.002,0.0025,0.0030,0.0035])

ax.set_yticklabels(['$2.0$','$2.5$','$3.0$','$3.5$'])

lgnd=ax.legend()

# Plot legend.
lgnd = plt.legend(loc="upper right", scatterpoints=1, fontsize=8)
lgnd.legendHandles[3]._sizes = [30]
lgnd.legendHandles[4]._sizes = [30]
lgnd.legendHandles[5]._sizes = [30]
lgnd.legendHandles[6]._sizes = [30]

ax.set(xlabel=r'$\sin^2 2\theta_{13}$', ylabel=r'$\Delta m^2_{ee} \hspace{0.5} [eV^{2}] \hspace{0.5} (x10^{-3})$')
ax.grid(True, linestyle='-')
ax.tick_params(labelcolor='black', labelsize=10, width=1)
#ig.savefig("reno_cov.eps")
plt.show()
```
